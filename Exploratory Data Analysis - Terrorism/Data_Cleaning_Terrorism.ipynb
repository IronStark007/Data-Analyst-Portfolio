## Preprocessing

```python
#checking info about the data
df.info() 
```

Some columns contains -9 as missing values therfore we have to change them to NA values

```python
df=df.replace(to_replace=-9,value=np.nan)
```

```python
#checking null values ratios
nulldf=df.isna().sum()/len(df)
nulldf=pd.DataFrame({"Colnames":df.columns,"Null values":nulldf.values})
nulldf=nulldf[nulldf['Null values']>=0.25]
nulldf=nulldf.transpose()
null_col=nulldf.loc['Colnames'].to_list()
print("Null values with more than 25% null values:\n",null_col)
print('Number of columns with more than 25% null values:',len(null_col))
```

Since all these 90 columns contains more than 30% null values we have to remove it

Also there are 19 columns which are not relevant so we will remove them too

```python
#dropping all the null columns & not important columns
not_imp=['eventid', 'country', 'region', 'specificity' , 'crit1' , 'crit2', 'crit3', 'multiple', 'attacktype1', 'targtype1', 'targsubtype1', 'natlty1', 'weaptype1', 'weapsubtype1', 'dbsource', 'INT_LOG', 'INT_IDEO', 'INT_MISC', 'INT_ANY']
null_col=null_col+not_imp
df.drop(null_col,axis=1,inplace=True)
```
